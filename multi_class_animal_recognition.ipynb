```python
dataset_path = "/kaggle/input/wild-animals/Animals/Cheetah"
dataset_path = "/kaggle/input/wild-animals/Animals/Lion"
dataset_path= "/kaggle/input/wild-animals/Animals/Tiger"

classes = os.listdir(dataset_path)
num_classes = len(classes)

# Display the number of classes along with names
print(f'Number of Classes: {num_classes}')
print(f'Classes: {classes}')
```

```python
plt.figure(figsize=(20,10))
for i, class_name in enumerate(classes[:10]):
  # class_path should be the directory containing images, not a specific image
  class_path = dataset_path  # Changed this line
  img_path = os.path.join(class_path, class_name)  # Changed this line
  img = plt.imread(img_path)

  plt.subplot(2, 5, i + 1)
  plt.imshow(img)
  plt.title(f'{class_name} \n shape: {img.shape}')
  plt.axis('off')
plt.show()
```
